```python
import pandas as pd
import numpy as np

# Load the datasets from the same folder
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

# Check the shape (number of rows and columns) of both datasets
print("Movies dataset shape:", movies.shape)
print("Credits dataset shape:", credits.shape)
```

```python
# Merge both datasets based on the 'title' column
movies = movies.merge(credits, on='title')

# Check the new shape after merging
# The number of columns should increase now
print("Merged dataset shape:", movies.shape)

# Display the first row to see all available columns
movies.head(1)
```

```python
# Select only the relevant columns for content-based recommendation
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]

# Verify the changes by printing the new head
movies.head()
```

```python
import ast

# Create a helper function to extract name tags from the messy column format
def convert(text):
    L = []
    # ast.literal_eval converts the string representation of a list into an actual list
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L

# Apply this helper function to the 'genres' column
movies['genres'] = movies['genres'].apply(convert)

# Apply the same helper function to the 'keywords' column
movies['keywords'] = movies['keywords'].apply(convert)

# Check the updated table to see the clean lists
movies.head()
```

```python
import ast

# Function to get the first 3 actors' names from the cast column
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

# Apply to cast column
movies['cast'] = movies['cast'].apply(convert3)

# Function to extract the Director's name from the crew column
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break # Once we find the director, we stop
    return L

# Apply to crew column
movies['crew'] = movies['crew'].apply(fetch_director)

# Verify the changes
movies.head()
```

```python
# Convert the overview string into a list of words
# We use lambda to split the sentence by spaces
movies['overview'] = movies['overview'].apply(lambda x:x.split() if isinstance(x, str) else [])

# Look at the data again
movies.head()
```

```python
# Function to remove spaces between words in a list
# Example: 'Sam Worthington' -> 'SamWorthington'
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])

# Let's see the result
movies.head()
```

```python
# Concatenate all the lists into a single column named 'tags'
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Create a new dataframe with only the columns we need
new_df = movies[['movie_id', 'title', 'tags']]

# Look at the new cleaned-up dataframe
new_df.head()
```

```python
# Create a fresh, independent copy of the dataframe to avoid warnings
new_df = movies[['movie_id', 'title', 'tags']].copy()

# Now perform the string operations safely
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

# Final check to see the clean text
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

# Initialize the PorterStemmer object
ps = PorterStemmer()

# Create a helper function to stem each word in a text block
def stem(text):
    y = []
    # Split the text into individual words
    for i in text.split():
        y.append(ps.stem(i)) # Stem each word (e.g., 'loving' -> 'love')
    return " ".join(y) # Join them back into a single string

# Apply the stem function to the 'tags' column
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer: max_features=5000 means we only want top 5000 words
# stop_words='english' will automatically remove common filler words
cv = CountVectorizer(max_features=5000, stop_words='english')

# Convert the text tags into a numeric matrix (Vectors)
vectors = cv.fit_transform(new_df['tags']).toarray()

# Check the shape of our vectors matrix
print("Vectors shape:", vectors.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity between all vectors
# This will compare every movie with every other movie
similarity = cosine_similarity(vectors)

# Check the shape of the similarity matrix
print("Similarity matrix shape:", similarity.shape)
```

```python
def recommend(movie):
    # Find the index of the movie given by the user
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    # Get the similarity scores for this movie
    distances = similarity[movie_index]
    
    # Sort the movies based on similarity scores in descending order
    # enumerate() helps us keep track of the original movie index after sorting
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Print the top 5 recommended movie titles
    print(f"Top 5 recommendations for '{movie}':")
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
import pickle

# 1. Save the movie dataframe as a dictionary
pickle.dump(new_df.to_dict(), open('movie_dict.pkl', 'wb'))

# 2. Convert similarity matrix to float16 to reduce size (Important Step)
# This will shrink the file size significantly without losing performance
similarity_light = similarity.astype('float16')

# 3. Save the smaller similarity matrix
pickle.dump(similarity_light, open('similarity.pkl', 'wb'))

print("Optimized Model files saved successfully!")
```
